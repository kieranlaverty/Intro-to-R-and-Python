Import data

```python
# Libraries 
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/lawre/OneDrive/Documents/Learning_R-main/Customers.csv")
print(df.head(10))
```

```python
df = pd.DataFrame(df)
df
```

```python
# Clean Data Drops any empty cells 
df = df.dropna()

print(df.to_string())
```

```python
# removes index code
df= df.dropna()
print(df)

# Drop a gender / any column 
#df = df.drop(columns=['Gender'])
#print(df)
```

```python
# Calcaute how much of Profession catagory count  
category_counts = df['Profession'].value_counts()
print(category_counts)
```

```python
# Create Pie chart with labels of entire data set
plt.figure(figsize=(8, 8))
category_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Customers by Country')
plt.ylabel('')  # Hides the default label
plt.show() 
```

```python
# Create Bar chart of all catagories 
plt.figure(figsize=(10, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette='viridis')
plt.title('Bar Chart of Customers by Country')
plt.ylabel('Number of Customers')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram for Age Distribution
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Customers', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Box Plot for Annual Income
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Annual Income ($)'], color='lightgreen')
plt.title('Box Plot of Annual Income', fontsize=14)
plt.xlabel('Annual Income (k$)', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Density Plot for Annual Income
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Annual Income ($)'], shade=True, color='purple')
plt.title('Density Plot of Annual Income', fontsize=14)
plt.xlabel('Annual Income (k$)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Bar Plot for Average Spending Score by Profession
plt.figure(figsize=(12, 6))
sns.barplot(x='Profession', y='Spending Score (1-100)', data=df, estimator=lambda x: x.mean(), palette='coolwarm')
plt.title('Average Spending Score by Profession', fontsize=14)
plt.xlabel('Profession', fontsize=12)
plt.ylabel('Average Spending Score', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Violin Plot for Spending Score by Profession
plt.figure(figsize=(12, 8))
sns.violinplot(x='Profession', y='Spending Score (1-100)', data=df, palette='muted')
plt.title('Spending Score by Profession', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
```

```python
# Pair Plot
sns.pairplot(df[['Age', 'Annual Income ($)', 'Spending Score (1-100)']], diag_kind='kde', markers='o', palette='Set1')
plt.suptitle('Pairwise Relationships between Age, Income, and Spending Score', y=1.02, fontsize=16)
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation = df.select_dtypes(include=['number']).corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap', fontsize=16)
plt.show()
```
